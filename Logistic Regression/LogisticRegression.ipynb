```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=10, n_informative=2, n_redundant=0, random_state=42, n_classes=2)
```

```python
df = pd.DataFrame(X, columns=[f'feature_{i}' for i in range(X.shape[1])])
df['target'] = y
```

```python
display(df.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('target', axis=1), df['target'], test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Logistic = LogisticRegression()

model = Logistic.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)
```

```python
print("Accuracy Score:\n", score)
print("\nConfusion Matrix:\n", cm)
print("\nClassification Report:\n", cr)
```

# Hyperparameter Tuning & Cross-Validation can be added here for better model performance

```python
model = LogisticRegression()
penalty = ['l1']
c_values = [0.01, 0.1, 1, 10, 100]
solver = ['liblinear', 'saga']
```

```python
param = dict(penalty=penalty, C=c_values, solver=solver)
```

```python
# Grid Search and Hyperparameter Tuning

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid = GridSearchCV(estimator=model, param_grid=param, cv=cv, n_jobs=-1, verbose=2, scoring='accuracy')
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.predict(X_test)
```

```python
# Rendomise search Search and Hyperparameter Tuning

from sklearn.model_selection import RandomizedSearchCV

model = RandomizedSearchCV(estimator=model, param_distributions=param, cv=cv, n_jobs=-1, verbose=2, scoring='accuracy')
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```
